```python
df['ingredients']=df['ingredients'].str.replace('*', '')
df['ingredients']=df['ingredients'].str.replace('(', '')
df['ingredients']=df['ingredients'].str.replace(')', '')
df
```

```python
df_ingr = df.explode('ingredients')
df_ingr.reset_index(drop=True)
```

```python
df_ingr['ingredients'].nunique()
```

```python
df_ingr['ingredients'] = df_ingr['ingredients'].apply(lambda x: x.replace('.', '') if x.endswith('.') else x)
df_ingr['ingredients'] = df_ingr['ingredients'].apply(lambda x: x.replace('[', '') if x.startswith('[') else x)
df_ingr['ingredients'] = df_ingr['ingredients'].apply(lambda x: x.replace('+', ''))
df_ingr['ingredients'] = df_ingr['ingredients'].apply(lambda x: x.replace('/', '') if x.startswith('/') else x)
df_ingr['ingredients'] = df_ingr['ingredients'].apply(lambda x: x.replace('-', '') if x.startswith('-') else x)
df_ingr['ingredients'] = df_ingr['ingredients'].apply(lambda x: x.strip())
df_ingr
```

```python
df_ingr.info()
```

```python
df_ingr.loc[(df_ingr['ingredients'].str.contains('water')==True) | (df_ingr['ingredients'].str.contains('aqua')==True), 'ingredients'] = 'water'
df_ingr.loc[(df_ingr['ingredients'].str.startswith('visit')==True), 'ingredients'] = 'no info'
df_ingr.loc[(df_ingr['ingredients'].str.contains('fragrance')==True) | (df_ingr['ingredients'].str.contains('parfum')==True), 'ingredients'] = 'fragrance'
```

```python
import re
df_ingr['ingredients'] = df_ingr['ingredients'].apply(lambda x: x.replace(':', ''))
df_ingr['ingredients'] = df_ingr['ingredients'].apply(lambda x: re.sub('\d', '', x) if '%' in x else x)
df_ingr['ingredients'] = df_ingr['ingredients'].apply(lambda x: x.replace('.', '') if '%' in x else x)
df_ingr
```

```python
df_ingr['ingredients'] = df_ingr['ingredients'].apply(lambda x: x.replace('%', ''))
df_ingr['ingredients'] = df_ingr['ingredients'].apply(lambda x: x.strip())
```

```python
unique_ingr = df_ingr['ingredients'].unique().tolist()
sorted(unique_ingr)[:20]
```

```python
import numpy as np
from fuzzywuzzy import process, fuzz

process.extract('1,2-hexanediol', unique_ingr, scorer=fuzz.token_sort_ratio)
```

```python
process.extract('zizyphus jujuba fruit extract', unique_ingr, scorer=fuzz.token_sort_ratio)
```

```python
def replace_matched_ingr(df, column, string_to_match, min_ratio=85):
    unique_ingr = df[column].unique()
    if(string_to_match not in unique_ingr):
        return
    matches = process.extract(string_to_match, unique_ingr, limit=10, scorer=fuzz.token_sort_ratio)

    close_matches = [matches[0] for matches in matches if matches[1] >= min_ratio]

    rows_with_matches = df[column].isin(close_matches or string_to_match)

    max = string_to_match
    for match in close_matches:
        if df[column].value_counts()[match] > df_ingr['ingredients'].value_counts()[max]: 
           max = match
    
    df.loc[rows_with_matches, column] = max
```

```python
for ingr in unique_ingr:
    replace_matched_ingr(df_ingr, 'ingredients', ingr)
```

```python
unique_ingr = df_ingr['ingredients'].unique()
ingr_ocurrence = {}
for ingr in unique_ingr:
    if df_ingr['ingredients'].value_counts()[ingr] >= 10:
        ingr_ocurrence[ingr] = df_ingr['ingredients'].value_counts()[ingr]
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(100,10))
plt.bar(*zip(*sorted(ingr_ocurrence.items())), align='edge', width=0.5)
plt.xticks(rotation = 90)
plt.show()
```

```python
df_ingr_cutoff = df_ingr[df_ingr['ingredients'].isin(ingr_ocurrence)]
df_ingr_cutoff.info()
df_ingr_cutoff['ingredients'].nunique()
```

```python
df_ingr_cutoff = df_ingr_cutoff.groupby(['Label', 'brand', 'name', 'price', 'rank', 'Combination', 'Dry', 'Normal', 'Oily', 'Sensitive'])['ingredients'].apply(list).reset_index()
df_ingr_cutoff['ingredients']
```
